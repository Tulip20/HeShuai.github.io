```python
# 导入 pyplot 模块，用于绑制图表
from matplotlib import pyplot as plt
# 导入 font_manager 模块，用于设置中文字体
from matplotlib import font_manager

# 设置中文字体为华文宋体，字号16，解决中文显示乱码问题
my_font = font_manager.FontProperties(fname='C:\\Windows\\Fonts\\STSONG.TTF', size=16)

# 2019年内地电影票房前6的电影名称
a = ['流浪地球', '疯狂的外星人', '飞驰人生', '大黄蜂', '熊出没·原始时代', '新喜剧之王']
# 对应的票房数据（单位：亿元）
b = [38.13, 19.85, 14.89, 11.36, 6.47, 5.93]

# 创建一个 20×8 英寸、分辨率为 80dpi 的画布
plt.figure(figsize=(20, 8), dpi=80)

# 绘制条形图：
#   range(len(a)) —— x 轴位置（0, 1, 2, ...）
#   b —— 每个条形的高度（票房数据）
#   width=0.3 —— 条形宽度
#   color='skyblue' —— 条形颜色为天蓝色
rects = plt.bar(range(len(a)), b, width=0.3, color='skyblue')

# 设置 x 轴刻度标签为电影名称，使用中文字体，旋转45度防止文字重叠
plt.xticks(range(len(a)), a, fontproperties=my_font, rotation=45)

# 在每个条形上方添加票房数值标注
for rect in rects:
    height = rect.get_height()          # 获取当前条形的高度
    plt.text(
        rect.get_x() + rect.get_width() / 2,  # x 坐标：条形水平居中位置
        height + 0.3,                          # y 坐标：条形顶部上方 0.3 处
        str(height),                           # 标注内容：票房数值
        ha="center"                            # 文字水平居中对齐
    )

# 显示图表
plt.show()
```

```python
# 导入 pyplot 模块，用于绑制图表
from matplotlib import pyplot as plt
# 导入 font_manager 模块，用于设置中文字体
from matplotlib import font_manager

# 3月份每天的温度数据（共31天）
y = [11,17,16,11,12,11,12,6,6,7,8,9,12,15,14,17,18,21,16,17,20,14,15,15,15,19,21,22,22,22,23]
# x 轴为日期，范围 1~31
x = range(1, 32)

# 设置画布大小为 20×8 英寸，分辨率 80dpi
plt.figure(figsize=(20, 8), dpi=80)

# 使用 scatter() 绘制散点图
#   x —— 横坐标（日期）
#   y —— 纵坐标（温度）
#   label='3月份' —— 图例中显示的名称
plt.scatter(x, y, label='3月份')

# 设置中文字体为华文宋体，字号10
my_font = font_manager.FontProperties(fname='C:\\Windows\\Fonts\\STSONG.TTF', size=10)

# 生成 x 轴刻度标签列表，如 ['3月1日', '3月2日', ..., '3月31日']
_xticks_labels = ['3月{}日'.format(i) for i in x]

# 设置 x 轴刻度：
#   x[::3] —— 每隔3天取一个刻度位置（1, 4, 7, 10, ...）
#   _xticks_labels[::3] —— 对应的标签（'3月1日', '3月4日', ...）
#   rotation=45 —— 标签旋转45度，防止重叠
plt.xticks(x[::3], _xticks_labels[::3], fontproperties=my_font, rotation=45)

# 设置 x 轴标题为"日期"
plt.xlabel('日期', fontproperties=my_font)
# 设置 y 轴标题为"温度"
plt.ylabel('温度', fontproperties=my_font)

# 显示图例（prop 参数指定图例使用的中文字体）
plt.legend(prop=my_font)

# 显示图表
plt.show()
```
